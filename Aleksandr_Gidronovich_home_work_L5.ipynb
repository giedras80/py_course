```python
def plus(a, b):
    return a + b

def minus(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: division by zero"
    return a / b

while True:
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    operation = input("Enter operation (+, -, *, /) or 'exit': ")
    
    if operation == "exit":
        print("Exiting calculator...")
        break

    if operation not in ["+", "-", "*", "/"]:
        print("Invalid operation")
        continue

    if operation == "+":
        result = plus(a, b)
        
    elif operation == "-":
        result = minus(a, b)
        
    elif operation == "*":
        result = multiply(a, b)
        
    elif operation == "/":
        result = divide(a, b)

    print(f"{a} {operation} {b} = {result}")
```

```python
# fuction factorial --> factorial of this number
```

```python
def myFactorial(n):
    if n < 0:
        return False
    if n <= 1:
        return 1
        
    targetVal = 1
    for i in range(2, n+1):
        targetVal *= i

    return targetVal 


def run():
    li = [-1, 0, 1, 2, 5]
    resList = [False, 1, 1, 2, 120]

    for number, result in zip(li, resList):
        if myFactorial(number) == result:
            print("OK")
        else:
            print("Failed")
            print("Expected: ", number, result)
            print("You've got: ", number, myFactorial(number))


run()            
```
