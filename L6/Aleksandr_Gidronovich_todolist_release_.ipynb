```python
def display_menu():
    print("\n=== TO-DO LIST MENU ===")
    print("1. Add task")
    print("2. Show tasks")
    print("3. Exit")


def display_task_list(tasks):
    if not tasks:
        print("Task list is empty.")
        return

    print("\n=== TASK LIST ===")
    for task in tasks:
        print(f"{task['id']}. {task['name']} (priority: {task['priority']})")


def enter_task(tasks):
    print("\nEnter tasks (type 'stop' to finish):")

    while True:
        name = input("Task name: ")
        if name.lower() == "stop":
            break

        priority = input("Priority (low/medium/high): ")

        task = {
            "id": len(tasks) + 1,
            "name": name,
            "priority": priority
        }

        tasks.append(task)
        print("Task added.")


def main():
    tasks = []

    while True:
        display_menu()
        choice = input("Choose option: ")

        if choice == "1":
            enter_task(tasks)

        elif choice == "2":
            display_task_list(tasks)

        elif choice == "3":
            print("Exiting program.")
            break

        else:
            print("Invalid choice. Try again.")


if __name__ == "__main__":
    main()
```
