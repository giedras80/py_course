```python
students = []
scores = []

while True:
    name = input("Enter student's name / leave empty & enter to finish.")
    if name == "":
        break

    try:
        score = int(input(f"Enter score for {name}: "))
    except ValueError:
        print("Invalid score. Try again.")
        continue

    students.append(name)
    scores.append(score)

if scores:
    average = sum(scores) / len(scores)
else:
    average = 0

    

print("\nStudents:")
print(*students, sep="\n")

print(f"\nAverage score: {average}")
```
